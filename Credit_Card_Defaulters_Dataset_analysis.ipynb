```python
#changing the values of target column ('default') from {1 : No default, 2 : Yes Default} to {0 : No default, 1 : Yes Default}
def change(line):
  if line == 1:
    return 0
  elif line == 2:
    return 1
```

```python
df['default'] = df['default'].apply(change)
df['default'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dummies = pd.get_dummies(df, drop_first = True)
```

```python
dummies
```

```python
target_col = 'default'
input_cols = dummies.drop(target_col, axis = 1).columns
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dummies[input_cols], dummies[target_col], test_size = 0.2, random_state = 1 )
```

Scaling features using StandardScaler and MinMaxScaler

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
scaler_std = StandardScaler()
X_train_std = scaler_std.fit_transform(X_train)
X_test_std = scaler_std.fit_transform(X_test)
```

```python
scaler_mm = MinMaxScaler()
X_train_mm = scaler_mm.fit_transform(X_train)
X_test_mm = scaler_mm.fit_transform(X_test)
```

**Logistic Regression Classifier**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.metrics import auc, roc_curve
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
print("Train score for Logistic Regression: {:.4f}".format(lr.score(X_train, y_train)))
print("Test score for Logistic Regression: {:.4f}".format(lr.score(X_test, y_test)))
```

```python
lr.fit(X_train_std, y_train)
print("Train score for StandardScaler scaled Logistic Regression: {:.4f}".format(lr.score(X_train_std, y_train)))
print("Test score for StandardScaler scaled Logistic Regression: {:.4f}".format(lr.score(X_test_std, y_test)))
```

```python
lr.fit(X_train_mm, y_train)
print("Train score for MinMaxScaler scaled Logistic Regression: {:.4f}".format(lr.score(X_train_mm, y_train)))
print("Test score for MinMaxScaler scaled Logistic Regression: {:.4f}".format(lr.score(X_test_mm, y_test)))
```

```python
def get_score(classifier, actual, predicted):
  accuracy = accuracy_score(actual, predicted)
  precision = precision_score(actual, predicted)
  recall = recall_score(actual, predicted)
  f1_score_ = f1_score(actual, predicted)
  tn, fp, fn, tp = confusion_matrix(actual, predicted).ravel()
  senstivity = tp/(tp + fn)
  specificity = tn/(tn + fp)
  fpr, tpr, thresholds = roc_curve(actual, predicted, pos_label = 1)
  auc_value = auc(fpr, tpr)

  return {'Classifier' : classifier,
          'Accuracy' : accuracy,
          'Precision' : precision,
          'Recall' : recall,
          'f1-score' :f1_score_,
          'Senstivity' : senstivity,
          'Specificity' : specificity,
          'AUC' : auc_value,
          'fpr value' : fpr,
          'tpr value' : tpr
          }
```

```python
y_pred_lr = lr.predict(X_test_mm)
lr_results = get_score('Logistic Regression Classifier', y_test, y_pred_lr)
print("Metrics for the Logistic regression classifier :", lr_results)
```

```python
results = ['Classifier', 'Accuracy', 'Senstivity', 'Specificity', 'f1-score', 'AUC']
df_results = pd.DataFrame(columns = results)
```

```python
df_results = df_results._append(lr_results, ignore_index=True)
```

```python
df_results
```

**Decision Tree Classifier**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
def dt_classifier(depth):
  dt = DecisionTreeClassifier(max_depth = depth, random_state = 2)
  dt.fit(X_train, y_train)
  print("Train score : {:.4f}".format(dt.score(X_train, y_train)))
  print("Test score : {:.4f}".format(dt.score(X_test, y_test)))
```

```python
for i in range(2, 20):
  print("Depth = ", i)
  dt_classifier(i)
  print("----------")
```

```python
#Using scaled features in classifier:
dt_std = DecisionTreeClassifier(max_depth = 14, random_state = 42)
dt_std.fit(X_train_std, y_train)
print("Decision Tree with Standard Scaler scaled features Train score : {:.4f}".format(dt_std.score(X_train_std, y_train)))
print("Decision Tree with Standard Scaler scaled features Test score : {:.4f}".format(dt_std.score(X_test_std, y_test)))
```

```python
dt_mm = DecisionTreeClassifier(max_depth = 14, random_state = 42)
dt_mm.fit(X_train_mm, y_train)
print("Decision Tree with MinMax Scaler scaled features Train score : {:.4f}".format(dt_mm.score(X_train_mm, y_train)))
print("Decision Tree with MinMax Scaler scaled features Test score : {:.4f}".format(dt_mm.score(X_test_mm, y_test)))
```

```python
#proceeding with MinMaxScaler features for result matrix:
y_pred_dt = dt_mm.predict(X_test_mm)
dt_results = get_score('Decision Tree Classifier', y_test, y_pred_dt)
```

```python
dt_results
```

```python
df_results = df_results._append(dt_results, ignore_index=True)
df_results
```

**K-Nearest Neighbors Classifier**

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
def knn_classifier(n):
  knn = KNeighborsClassifier(n_neighbors = n, metric = 'minkowski')
  knn.fit(X_train, y_train)
  print("Train score : {:.4f}".format(knn.score(X_train, y_train)))
  print("Test score : {:.4f}".format(knn.score(X_test, y_test)))
```

```python
for neighbors in range(2, 10):
  print('Neighbors = ', neighbors)
  knn_classifier(neighbors)
  print("------")
```

The Best test score for KNN classifier is for n = 8

```python
#Using scaled features in classifier:
knn_std = KNeighborsClassifier(n_neighbors = 8)
knn_std.fit(X_train_std, y_train)
print("KNN with Standard Scaler scaled features Train score : {:.4f}".format(knn_std.score(X_train_std, y_train)))
print("KNN with Standard Scaler scaled features Test score : {:.4f}".format(knn_std.score(X_test_std, y_test)))
```

```python
knn_mm = KNeighborsClassifier(n_neighbors = 8)
knn_mm.fit(X_train_mm, y_train)
print("KNN with MinMax Scaler scaled features Train score : {:.4f}".format(knn_mm.score(X_train_mm, y_train)))
print("KNN with MinMax Scaler scaled features Test score : {:.4f}".format(knn_mm.score(X_test_mm, y_test)))
```

The Test score has improved with scaling of features which shows that KNN performs better with scaled features  

```python
y_pred_knn = knn_std.predict(X_test_std)
knn_results = get_score('K-Nearest Neighbors Classifier', y_test, y_pred_knn)
```

```python
df_results = df_results._append(knn_results, ignore_index = True)
df_results
```
